```python
#Step1.
# matplotlib.pyplotメソッドを定義するコード
import matplotlib.pyplot as plt
# numpyをインポートするコード
import numpy as np

#Step2.
# 6個のグラフを表示する領域を作成
fig, axes = plt.subplots(2,3,tight_layout=True)

#Step3.
#各領域に、折れ線グラフ、sin関数、ヒストグラム、散布図のデータを表示するコードを記述する

#折れ線グラフ
# y = [1, 2, -5, 2]
# plt.plot(y)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

axes[0,1] .plot(x, y, linewidth=2.0)

axes[0,1] .set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x = np.random .normal(15, 5, 2000)
axes[0, 2].set_xlim(-10, 40)
# ヒストグラムをプロット
axes[0, 2].hist(x, bins=30)

# 散布図

# データを生成する
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# プロットする
axes[1, 0].scatter(x, y)

# Step4. グラフを表示させる行「plt.show()」を書く
plt.show()


```
